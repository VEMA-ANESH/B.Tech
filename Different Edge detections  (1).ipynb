```python
# Import necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imshow
from skimage.feature import blob_dog, blob_log, blob_doh
```

```python
# Function to display images using matplotlib
def display_image(image, title):
    plt.imshow(image, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()

```

```python
# Read an example image

original_image = cv2.imread('black moon.jpeg')

```

```python
# Display the original image
display_image(original_image, 'Original Image')

```

```python
# Apply different edge detection techniques
# 1. Canny Edge Detection
canny_edges = cv2.Canny(original_image, 50, 150)
display_image(canny_edges, 'Canny Edge Detection')

```

```python
# 2. Sobel Edge Detection
sobelx = cv2.Sobel(original_image, cv2.CV_64F, 1, 0, ksize=5)
sobely = cv2.Sobel(original_image, cv2.CV_64F, 0, 1, ksize=5)
sobel_edges = np.sqrt(sobelx**2 + sobely**2)
display_image(sobel_edges, 'Sobel Edge Detection')

```

```python
# 3. Laplacian Edge Detection
laplacian_edges = cv2.Laplacian(original_image, cv2.CV_64F)
display_image(laplacian_edges, 'Laplacian Edge Detection')


```

```python
# 4. Scharr Edge Detection
scharrx = cv2.Scharr(original_image, cv2.CV_64F, 1, 0)
scharry = cv2.Scharr(original_image, cv2.CV_64F, 0, 1)
scharr_edges = np.sqrt(scharrx**2 + scharry**2)
display_image(scharr_edges, 'Scharr Edge Detection')
```
